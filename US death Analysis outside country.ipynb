```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("2002-10-2018-06.csv")
data.head(5)
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
plt.figure(figsize=(15,7))
sns.countplot(x='Cause of Death',data=data,order=pd.value_counts(data['Cause of Death']).iloc[:10].index)
plt.title("Top 10 Cause of death")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(15,7))
sns.countplot(x='City',data=data,order=pd.value_counts(data['City']).iloc[:10].index)
plt.title("Top 10 City of death")
plt.xticks(rotation=90)
plt.show()
```

```python
data['year'] = pd.DatetimeIndex(data['Date']).year
```

```python
data.head()
```

```python
data = data.dropna()
data = data.reset_index(drop = True)
```

```python
data.head()
```

```python
data['year'] = data['year'].apply(np.int64)
```

```python
data.head()
```

```python
#sns.countplot(x ='sex', hue = "smoker", data = df)
plt.figure(figsize=(15,7))
sns.countplot(x='year',data=data,order=pd.value_counts(data['year']).iloc[:10].index)
plt.title("Top 10 years of death")
plt.xticks(rotation=90)
plt.show()
```

```python
#sns.countplot(x ='sex', hue = "smoker", data = df)
plt.figure(figsize=(15,7))
sns.countplot(x='City',hue='year',data=data,order=pd.value_counts(data['City']).iloc[:10].index)
plt.title("Top 10 years of death wrt city")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(15,7))
sns.countplot(x='Cause of Death',hue='year',data=data,order=pd.value_counts(data['Cause of Death']).iloc[:10].index)
plt.title("Top 10 years of death wrt Cause of Death")
plt.xticks(rotation=90)
plt.show()
```

```python
data['month'] = pd.DatetimeIndex(data['Date']).month
```

```python
data.head()
```

```python
plt.figure(figsize=(15,7))
sns.countplot(x='month',data=data,order=pd.value_counts(data['month']).iloc[:10].index)
plt.title("Top 10 months of death from 2002 to 2018")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(15,7))
sns.countplot(x='Cause of Death',hue='month',data=data,order=pd.value_counts(data['Cause of Death']).iloc[:10].index)
plt.title("12 months of death wrt Cause of Death")
plt.xticks(rotation=90)
plt.show()
```

```python
#sns.countplot(x ='sex', hue = "smoker", data = df)
plt.figure(figsize=(15,7))
sns.countplot(x='City',hue='month',data=data,order=pd.value_counts(data['City']).iloc[:10].index)
plt.title("12 months of death wrt city")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(15,7))
sns.countplot(x='year',hue='month',data=data,order=pd.value_counts(data['year']).iloc[:10].index)
plt.title("top 10 years wrt  months of death ")
plt.xticks(rotation=90)
plt.show()
```
